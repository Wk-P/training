```python
# import 
from pathlib import Path
```

```python
# import data from .xlsx file
import pandas as pd

def import_data(file_path):
    # Read the Excel file into a DataFrame
    df = pd.read_excel(file_path)
    return df
```

```python
xls_file_path = Path.cwd().parent.parent / "training_data" / "RANDclient_org-L1-G200-RB48498-DTFriSep132216422024.xlsx"

print(xls_file_path)  # Replace with your actual file path
```

```python
# read data

data_df = import_data(xls_file_path)

print(data_df.head())  # Display the first few rows of the DataFrame

request_numbers = data_df['request_num'].tolist()
processed_time = data_df['processed_time'].tolist()
```

```python
# write to .json file
import json

output_data = {
    "request_number": request_numbers, 
    "real_process_time": processed_time
}

with open(Path.cwd() / '20251112_merge_data.json', 'w') as json_file:
    json.dump(output_data, json_file, indent=4) 
```
